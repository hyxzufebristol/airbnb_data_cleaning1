```python
# remove '$' of 'price'
def format_price(price):
    return(float(price.replace('$','').replace(',','')))

df_listings['price_$'] = df_listings['price'].apply(format_price) 
df_listings[['price','price_$']].head()
```

```python
# (homes' names of 100 highest profit_per_month)

df_listings['profit_per_month'] = df_listings['price_$'] * df_listings['reviews_per_month']
df_listings = df_listings.sort_values(by=['profit_per_month'],ascending=False)
df_listings[['name','host_name','price_$','number_of_reviews','reviews_per_month','profit_per_month']].head(100)
```

```python
# (homes' names of 100 highest profit_per_month)

df_listings['profit_sum'] = df_listings['price'] * df_listings['number_of_reviews']
df_listings = df_listings.sort_values(by=['profit_sum'],ascending=False)
df_listings[['name','host_name','price','number_of_reviews','reviews_per_month','profit_per_month','profit_sum']].head(100)
```
